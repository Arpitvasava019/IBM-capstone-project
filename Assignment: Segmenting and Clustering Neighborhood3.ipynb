```python
#importing libraries
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
#scrapping data from given wikipedia link
wiki_link = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
wiki_page = requests.get(wiki_link).text

#parse XML using Beautiful soup

soup = BeautifulSoup(wiki_page,'xml')
print(soup)
```

# Extracting data from XML file.

```python
#extracting table content
table = soup.find('table')
Postcode_data = []
Borough_data = []
Neighbourhood_data = [] 

#modifying data into table format from raw tble format

for row in table.find_all('tr'):
    
    count=1
    Postcode=-1
    Borough=-1
    Neighbourhood=-1
    
    for data in row.find_all('td'):
        if count == 1:
            Postcode = data.text
        if count == 2:
            Borough=data.text
            Borough_tag=data.find('a')
            
        if count==3:
        
            Neighbourhood=str(data.text).strip()
            Neighbourhood_tag=data.find('a')
            
        count +=1

        if (Postcode == 'Not assigned' or Borough == 'Not assigned' or Neighbourhood == 'Not assigned'):
            
            continue
            
           
    try:
        if ((Borough_tag is None) or (Neighbourhood_tag is None)):
            
            continue
            
            
    except:
        
        pass
    
    if(Postcode==-1 or Borough==-1 or Neighbourhood==-1):

        continue
        
        
     
    Postcode_data.append(Postcode)
    Borough_data.append(Borough)
    Neighbourhood_data.append(Neighbourhood)
```

# Combining postal codes which are same

```python
unique_code = set(Postcode_data)
print('num of unique Postal codes:', len(unique_code))
Postcode_tbl      = [ ]
Borough_tbl       = [ ]
Neighbourhood_tbl = [ ]


for unique_postcode in unique_code:
    post_var = ''; boro_var = ''; neigh_var = ''; 
    for postcode_idx, postcode_element in enumerate(Postcode_data):
        if unique_postcode == postcode_element:
            post_var = postcode_element;
            boro_var = Borough_data[postcode_idx]
            if neigh_var == '': 
                neigh_var = Neighbourhood_data[postcode_idx]
            else:
                neigh_var = neigh_var + ', ' + Neighbourhood_data[postcode_idx]
    Postcode_tbl.append(post_var)
    Borough_tbl.append(boro_var)
    Neighbourhood_tbl.append(neigh_var)
```

# Creating pandas dataframe

```python
toronto = {'Postcode':Postcode_tbl, 'Borough':Borough_tbl, 'Neighbourhood':Neighbourhood_tbl}
df_toronto = pd.DataFrame.from_dict(toronto)
df_toronto.to_csv('toronto.csv')
df_toronto.head(14)
```
